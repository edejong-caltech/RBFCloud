```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Import data from constant kernel
ckern_IC1 = pickle.load(open("constant_IC1_errors.pkl","rb"))
ckern_IC2 = pickle.load(open("constant_IC2_errors.pkl","rb"))
ckern_bi = pickle.load(open("constant_bimodal_errors.pkl","rb"))
golovin_IC1 = pickle.load(open("golovin_IC1_errors.pkl","rb"))
golovin_IC2 = pickle.load(open("golovin_IC2_errors.pkl","rb"))
golovin_bi = pickle.load(open("golovin_bimodal_errors.pkl","rb"))
hydro_IC1 = pickle.load(open("geometric_IC1_errors.pkl","rb"))
hydro_IC2 = pickle.load(open("geometric_IC2_errors.pkl","rb"))
hydro_bi = pickle.load(open("geometric_bimodal_errors.pkl","rb"))
```

```python
fig, ax = plt.subplots(nrows = 3, sharex=True, figsize=(5, 12))
#plt.rcParams.update({'font.size':11})
colors = ['#000000','#91bfdb','#4575b4','#fc8d59','#d7191c']

ind = np.arange(3)
width = 0.2


ax[0].set_title('EXP Initial Condition')
rbf8_IC2 = [ckern_IC2[0],golovin_IC2[0],hydro_IC2[0]]
rbf16_IC2 = [ckern_IC2[1],golovin_IC2[1],hydro_IC2[1]]
mom_IC2 = [ckern_IC2[2],golovin_IC2[2],hydro_IC2[2]]
bott_IC2 = [ckern_IC2[3],golovin_IC2[3],hydro_IC2[3]]
ax[0].bar(ind, rbf8_IC2, width, color=colors[1], label='Collocation n_bf = 8')
ax[0].bar(ind + width, rbf16_IC2, width,color=colors[2], label='Collocation n_bf = 16')
ax[0].bar(ind + 2*width, bott_IC2, width, color=colors[3], label='Flux method n_bin = 32')
ax[0].bar(ind + 3*width, mom_IC2, width, color=colors[4], label='MOM (2 moments)')

ax[1].set_title('GAM Initial Condition')
rbf8_IC1 = [ckern_IC1[0],golovin_IC1[0],hydro_IC1[0]]
rbf16_IC1 = [ckern_IC1[1],golovin_IC1[1],hydro_IC1[1]]
mom_IC1 = [ckern_IC1[2],golovin_IC1[2],hydro_IC1[2]]
bott_IC1 = [ckern_IC1[3],golovin_IC1[3],hydro_IC1[3]]
ax[1].bar(ind, rbf8_IC1, width, color=colors[1], label='Collocation n_bf = 8')
ax[1].bar(ind + width, rbf16_IC1, width,color=colors[2], label='Collocation n_bf = 16')
ax[1].bar(ind + 2*width, bott_IC1, width, color=colors[3], label='Flux method n_bin = 32')
ax[1].bar(ind + 3*width, mom_IC1, width, color=colors[4], label='MOM (3 moments)')

ax[2].set_title('BIM Initial Condition')
rbf8_bi = [ckern_bi[0],golovin_bi[0],hydro_bi[0]]
rbf16_bi = [ckern_bi[1],golovin_bi[1],hydro_bi[1]]
mom_bi = [ckern_bi[2],golovin_bi[2],hydro_bi[2]]
bott_bi = [ckern_bi[3],golovin_bi[3],hydro_bi[3]]
ax[2].bar(ind, rbf8_bi, width, color=colors[1], label='Collocation n_bf = 8')
ax[2].bar(ind + width, rbf16_bi, width,color=colors[2], label='Collocation n_bf = 16')
ax[2].bar(ind + 2*width, bott_bi, width, color=colors[3], label='Flux method n_bin = 32')
ax[2].bar(ind + 3*width, mom_bi, width, color=colors[4], label='MOM (2 moments)')


ax[0].set_ylabel('Relative $L_2$ error')
ax[1].set_ylabel('Relative $L_2$ error')
ax[2].set_ylabel('Relative $L_2$ error')
ax[0].set_ylim([0.0, 1.4])
ax[1].set_ylim([0.0, 1.4])
ax[2].set_ylim([0.0, 1.4])
plt.xticks(ind + width / 2, ('Constant K', 'Golovin', 'Hydrodynamic'))
ax[0].legend(loc='best')
ax[1].legend(loc='best')
ax[2].legend(loc='best')
plt.tight_layout()
plt.savefig("errors_compared_allICs.pdf")
```
